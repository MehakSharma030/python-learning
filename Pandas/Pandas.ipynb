# create series from dictionary

```python
fruit_protein = {
    "avacado":2.4 ,
    "guava":4.1,
    "oranges":3.2,
    "apples":8.1,
    "kiwi":4.2,
    "pomegranate":6.4,
    "mango":6.3,
    "cherries":7.2,
    "blackberries":2.2,
    "banana":6.4
}
s2 = pd.Series(fruit_protein , name = "protein")
s2
```

```python
s2 > 3
```

```python
s2[s2 > 3]
```

```python
# LOGICAL OPERATORS

# AND
s2 [ (s2 > 3) & (s2 < 8) ]
```

```python
# OR
s2 [ (s2 > 3) | (s2 < 8) ]
```

```python
# NOT
s2 [ ~ (s2 > 3) ]
```

```python
# modifying the series
s2['kiwi']=5.3
s2
```

data cleaning

```python
df.isnull().sum()
```

```python
# to drop the rows having null values
df.dropna()
# or df.dropna(how = 'any')  -> any row that had any null value
# df.drop(how = 'all')  -> delte only those rows where all values are nulll
```

```python
df
```

```python
# replace na values with some other values
df.fillna(0)
```

```python
# to fill the ( na ) value with mean
df["Age"].fillna(  df["Age"].mean() )
```

```python
# to fill the ( na ) value with median
df["Salary"].fillna(  df["Salary"].median() )
```

```python
# fillna -> forward fill and backward fill
# forward fill :  top to botton (top value will be replaced with the just next missing value)
# backward fill : bottom to top (bottom value will be replaced with the just bottom past value)
```

```python
# backward fill
df["Age"].fillna(method = "bfill" )
```

```python
# forward fill
df["Age"].fillna(method = "ffill" )
```

```python
# to change the name of a student
df["Name"].replace( "Charlie","Rose")
```

```python
# if to change the name in the original dataset also
df["Name"] = df["Name"].replace( "charlie","Rose")
```

```python
df
```

dealing with duplicate values

```python
# duplicated()  -> it will return the second duplicated value
df_dup = df[ df.duplicated() ]
df_dup
```

```python
df_dup = df[ df.duplicated( keep = 'first' ) ] # by default it takes the value as keep = 'first
df_dup
```

```python
df_dup = df[ df.duplicated( keep = 'last' ) ]  # it works from bottom to top
df_dup
```

```python
# to remove the duplicate values
df = df.drop_duplicates()
```

```python
df
```

INVALID VALUES

```python
df["Promoted_Salary"] = df["Promoted_Salary"].apply( lambda x : x/10 if x >= 650000 else x)
```

```python
df
```

```python
# if we have a string input of name ad first name and last name then it can be split into two coloumns namely FIRST NAME and LAST NAME using split function of string
name = "alice_fernedes"
df[ ['first_name','last_name'] ] = df["name"].str.split("_")
```

apply and lambda

```python
# these help the change the single entries in a string
def multiplying_age(x):
  return x*2

df["Age"] = df["Age"].apply( multiplying_age )

# or it can also be written as below

# df["Age"] = df["Age"].apply(lambda x : x*2)
```

```python
df
```

```python
# converting ages back to the original values
df["Age"] = df["Age"].apply(lambda x : x/2)
```

```python
df
```

JOINS AND MERGES

```python
# JOINS = LEFT JOIN , RIGHT JOIN , OUTER JOIN , INNER JOIN
department_inf0 = {
    "Dept":['HR','IT','FINANCE'],
    "Location":['NEW YORK','SAN FRANCISCO','CHICAGO'],
    "Manager":['LAIRA','STEVE','NINA']
}
df2 = pd.DataFrame(department_inf0)
```

```python
df2
```

```python
# to combine df and df2 -> CONCAT function
pd.concat( [df,df2] )
```

```python
# if to merge the dataframes coloumn wise
pd.concat( [df,df2] , axis=1)
```

```python
# MERGE -> TO MERGE THE DATAFRAMES ON A COMMON COLOUMN
pd.merge( df , df2 ,on = 'Dept' )
```
