# **program to find the largest and smallest element in the last**

```python
list = [10,20,30,40,50,60,70,80,90]
print("list : ",list)

minm = list[0]
maxm = list[0]

for i in range(0,len(list)):
  if(list[i] > maxm):
    maxm = list[i]
  if(list[i] < minm):
    minm = list[i]

print("maximum elem : ",maxm)
print("minimum elem : ",minm)
```

# **program to remove duplicates in a list**

```python
list = [10,20,30,20,30,60,10,20,10]
print("original list : ",list)

new_list = []
for i in range(0,len(list)):
  if(list[i] not in new_list):
    new_list.append(list[i])

print("after duplicates removal : ",new_list)
```

# **program to sort list without using sort**

```python
list = [40,50,60,70,80,90,10,20,30]
print("original list : ",list)

for i in range(0,len(list)):

  for j in range(i+1,len(list)):

    if(list[i] > list[j]):
      temp = list[i]
      list[i] = list[j]
      list[j] = temp

print("sorted list  :  ",list)
```

## **program to find frequency of each element**

```python
list = [40,50,60,70,80,90,10,20,30]
print("list : ",list)

freq = {}

for item in range(0,len(list)):
  if(item in freq):
    freq[item] += 1
  else:
    freq[item] = 1

print("freq of each element : ",freq)
```

# **program to find common elements in a list**

```python
list1 = [40,50,60,70,80,90,10,20,30]
print("list1:", list1)

list2 = [1,5,40,50,10,30,5,20,6,7,9]
print("list2:", list2)

result = []

# sort both lists
list1.sort()
list2.sort()

i, j = 0, 0

while (i < len(list1) and j < len(list2)):
    if (list1[i] == list2[j]):
        result.append(list1[i])
        i += 1
        j += 1
    elif (list1[i] < list2[j]):
        i += 1
    else:
        j += 1

print("common elements in both lists are:", result)
```

## **program to find prime numbers in a list**

```python
list1 = [2,40,60,81,93,41,90]
print("list : ",list1)

primes = []

for i in list1:
  is_prime = True
  for j in range(2 ,i-1):
    if(i % j == 0):
      is_prime = False
      break
  if(is_prime):
    primes.append(i)

print("primes : ",primes)
```

# **program to merge two lists**

```python
list1 = [1, 2, 3]
print("list1 : ",list1)

list2 = [2 ,4 , 5, 6]
print("list2 : ",list2)

merged = []

for i in (list1+list2):
  if i not in merged:
    merged.append(i)

print("merged : ",merged)
```
