# **BREAK**

```python
# 1. Stop loop when number becomes 5

for i in range(1,11):

    if i == 5:
        break

    print(i)
```

```python
# 2. Search an element in list

arr = [10,20,30,40,50]

for i in arr:

    if i == 30:
        print("Element Found")
        break
```

```python
# 4. Infinite loop stopped using break

i = 1

while True:

    print(i)

    if i == 5:
        break

    i += 1
```

```python
# 4. Break when character is found

word = "PYTHON"

for ch in word:

    if ch == "H":
        break

    print(ch)
```

```python
# 5. Stop printing when negative number appears

arr = [10, 20, 30, -1, 40, 50]

for i in arr:

    if i < 0:
        break

    print(i)
```

# **CONTINUE**

```python
# 1. Skip number 5

for i in range(1,11):

    if i == 5:
        continue

    print(i)
```

```python
# 2. Print only odd numbers

for i in range(1,11):

    if i % 2 == 0:
        continue

    print(i)
```

```python
# 3. Skip vowels in a string

word = "PYTHONPROGRAM"

for ch in word:

    if ch in "AEIOU":
        continue

    print(ch)
```

```python
# 4. Skip negative numbers

arr = [10, -2, 30, -5, 40]

for i in arr:

    if i < 0:
        continue

    print(i)
```

```python
# 5. Skip multiples of 3

for i in range(1,16):

    if i % 3 == 0:
        continue

    print(i)
```
