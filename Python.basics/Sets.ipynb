# set operations

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# union
print("union : ",s1 | s2)

# intersection
print("intersection : ",s1 & s2)

# differnce
print("Difference : ",s1-s2)

# symmetric differnce ---> A△B = (A−B) ∪ (B−A)
print("symmetric differnece : ",s1 ^ s2)
```

# set functions

```python
# add() --> adds a single element at the end of a set
s1 ={1,2,3,4,5}
print(s1)
s1.add(10)
print(s1)
```

```python
# len() --> find length of a set
s1 ={1,2,3,4,5}
print("S1 : ",s1)

print( "length : ",len(s1) )
```

```python
# max() --> find maximum element of a set
s1 ={1,2,3,4,5}
print("S1 : ",s1)

print("max_elem : ",max(s1))
```

```python
# min() --> find minimum element of a set
s1 ={1,2,3,4,5}
print("S1 : ",s1)

print("min_elem : ",min(s1))
```

```python
# sum() --> find sum of elements of a set
s1 ={1,2,3,4,5}
print("S1 : ",s1)

print("sum : ",sum(s1))
```

```python
# update() --> add multiple elements to the end of a set
s1 ={1,2,3,4,5}
print("S1 : ",s1)
s1.update([10,11,12])
print(s1)
```

```python
# remove() --> removes an element ; if not found gives error
s1 ={1,2,3,4,5}
print("S1 : ",s1)
s1.remove(3)
print("S1 : ",s1)
```

```python
# discard() --> removes an element ; if not found doesnot retur an error
s1 ={1,2,3,4,5}
print("S1 : ",s1)
s1.discard(3)
print("S1 : ",s1)
```

```python
# pop() --> removes and returns a random element
s1 ={1,2,3,4,5}
print("S1 : ",s1)
s1.pop()
```

```python
# clear() -> removes all elements from the set
s1 ={1,2,3,4,5}
print("S1 : ",s1)

s1.clear()
print("S1 : ",s1)
```

```python
# copy() --> creates a shallow copy of the set
s1 ={1,2,3,4,5}
print("S1 : ",s1)
s2 = s1.copy()
print("s2 : ",s2)
```

```python
# union() --> return or (+) of two sets
s1 ={1,2,3,4,5}
print("S1 : ",s1)
s2 ={4,5,6,7,8,9,10}
print("S2 : ",s2)

print("union : ",s1.union(s2) )
```

```python
# intersection() --> returns and(*) of two sets
s1 ={1,2,3,4,5}
print("S1 : ",s1)
s2 ={4,5,6,7,8,9,10}
print("S2 : ",s2)

print("intersection : ",s1.intersection(s2))
```

```python
# differnce() --> returns the difference of two sets
s1 ={1,2,3,4,5}
print("S1 : ",s1)
s2 ={4,5,6,7,8,9,10}
print("S2 : ",s2)

print("difference (s1-s2) : ",s1.difference(s2))
print("difference (s2-s1) : ",s2.difference(s1))
```

```python
# symmetric_difference() --> calcualtes (A - B) U (B - A)
s1 ={1,2,3,4,5}
print("S1 : ",s1)
s2 ={4,5,6,7,8,9,10}
print("S2 : ",s2)

print("symmetric difference : ",s1.symmetric_difference(s2))
```

```python
# issubset()
s1 = {1, 2, 3, 4, 5}
print("S1 :", s1)

s2 = {1, 2, 3}
print("S2 :", s2)

if s2.issubset(s1):
    print("s2 is a subset of s1")
else:
    print("s2 is not a subset of s1")
```

```python
# issuperset()
s1 = {1, 2, 3, 4, 5}
print("S1 :", s1)

s2 = {1, 2, 3}
print("S2 :", s2)

if s1.issuperset(s2):
    print("s1 is a superset of s2")
else:
    print("s1 is not a superset of s2")
```

```python
# isdisjoint() --> that have no common element in between them
s1 = {1, 2, 3, 4, 5}
print("S1 :", s1)

s2 = {1, 2, 3}
print("S2 :", s2)

if s1.isdisjoint(s2):
    print("s1 and s2 are disjoint sets")
else:
    print("s1 and s2 are not disjoint sets")
```
