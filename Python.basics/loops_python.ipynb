# **FOR LOOP**

```python
# 1. Print numbers from 1 to 5

for i in range(1,6):
    print(i)
```

```python
# 2. Print even numbers

for i in range(2,11,2):
    print(i)

# 3. Print odd numbers

for i in range(1,10,2):
    print(i)
```

```python
# 4. Sum of numbers from 1 to 5

sum = 0

for i in range(1,6):
    sum = sum + i

print("Sum =", sum)
```

```python
# 5. Multiplication table of 5

for i in range(1,11):
    print("5 x", i, "=", 5*i)
```

```python
# 6. Print characters of a string

name = "Python"

for ch in name:
    print(ch)
```

```python
# 7. Find square of numbers

for i in range(1,6):
    print(i*i)
```

```python
# 8. Nested for loop

for i in range(1,4):
    for j in range(1,4):
        print(i, j)
```

```python
# 9 Star Pattern

for i in range(1,6):
    print("*" * i)
```

```python
# 10. Count total vowels in a string

name = "education"
count = 0

for ch in name:
    if ch in "aeiou":
        count += 1

print("Vowels =", count)
```

```python
# 11. Find factorial of a number

num = 5
fact = 1

for i in range(1, num+1):
    fact = fact * i

print("Factorial =", fact)
```

```python
# 12. Find largest number in list

arr = [12,45,7,89,23]

largest = arr[0]

for i in arr:
    if i > largest:
        largest = i

print("Largest =", largest)
```

```python
# 13 Print ASCII values of characters

word = "ABC"

for ch in word:
    print(ch, "=", ord(ch))
```

```python
# 14. Reverse a string

word = "Python"
rev = ""

for ch in word:
    rev = ch + rev

print("Reverse =", rev)
```

```python
# 15. Count digits in a number

num = 12345
count = 0

for i in str(num):
    count += 1

print("Digits =", count)
```

# **WHILE LOOP**

```python
# 1. Print numbers from 1 to 5

i = 1

while i <= 5:
    print(i)
    i += 1
```

```python
# 2. Print even numbers

i = 2

while i <= 10:
    print(i)
    i += 2
```

```python
# 3. Print odd numbers

i = 1

while i <= 10:
    print(i)
    i += 2
```

```python
# 4. Sum of first 5 numbers

i = 1
sum = 0

while i <= 5:
    sum += i
    i += 1

print("Sum =", sum)
```

```python
# 5. Multiplication table of 5

i = 1

while i <= 10:
    print("5 x", i, "=", 5*i)
    i += 1
```

```python
# 6. Reverse counting

i = 10

while i >= 1:
    print(i)
    i -= 1
```

```python
# 7. Print characters of string

name = "Python"
i = 0

while i < len(name):
    print(name[i])
    i += 1
```

```python
# 8. Factorial of a number

num = 5
fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print("Factorial =", fact)
```

```python
# 9. Count digits in a number

num = 12345
count = 0

while num > 0:
    count += 1
    num = num // 10

print("Digits =", count)
```

```python
# 10. Reverse a number

num = 1234
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reverse =", rev)
```

```python
# 11. Palindrome number check

num = 121
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp = temp // 10

if num == rev:
    print("Palindrome")

else:
    print("Not Palindrome")
```

```python
# 12. Print squares of numbers

i = 1

while i <= 5:
    print(i*i)
    i += 1
```

```python
# 13. Find largest number in list

arr = [12,45,7,89,23]

i = 0
largest = arr[0]

while i < len(arr):

    if arr[i] > largest:
        largest = arr[i]

    i += 1

print("Largest =", largest)
```

```python
# 14. Fibonacci Series

a = 0
b = 1
i = 1

while i <= 10:
    print(a)

    c = a + b
    a = b
    b = c

    i += 1
```

```python
# 15. Prime number check

num = 7
i = 1
count = 0

while i <= num:

    if num % i == 0:
        count += 1

    i += 1

if count == 2:
    print("Prime Number")

else:
    print("Not Prime")
```
