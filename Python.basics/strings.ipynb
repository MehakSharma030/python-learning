# **string functions**

```python
str1 = "mehak  sharma"

print("Str1 : ",str1)

print()

# to find length
print("length str1 : ",len(str1) )

# to convert to upper case
print("to lower : ",str1.lower())

# to convert to lower case
print("to upper : ",str1.upper())

# title
print("title : ",str1.title() )

# capitalise
print("capitalise : ",str1.capitalize())

# count the number of occurences of a letter
print("count(a) : ",str1.count("a"))

print("count(meh) : ",str1.count("meh"))
print()

#strip -> remove extra spaces from left and right
str2 = " python code "
print("str2 : ",str2)
print("length : ",len(str2))
print()

print("Strip -> ",str2.strip() )

#lstrip( remove spaces from left ) and rstrip( remove spaces from right )
print("lstrip -> ",str2.lstrip() )
print("rstrip -> ",str2.rstrip())

# replace function
print("Str1.replace(m,M) : ",str1.replace("m", "M") )

# find -> finds first index of substring (returns -1 if not found)
print("str1.find(a) : ",str1.find("a"))

# index() -> Same as find() but gives error if not found
print("str1.index(a) : ",str1.index("a"))

# startswith()
print("str1.startswith(m) : ",str1.startswith("m") )
print("str1.startswith(M) : ",str1.startswith("M") )

# endswith()
print("str1.endswith(a ) : ",str1.endswith("a") )
print("str1.endswith(A) : ",str1.endswith("A") )


# join
str3 = [" my " , "code ", " is " , " in "," python "]

print(" " ".join(str3) : "," ".join(str3))
print(" ,.join(str3) : "," , ".join(str3))

# split
print("str1.split() : ",str1.split())

#
```

# **program to count frequency of each character in a string**


```python
str = " i am a coder "
freq = {}
for char in str:
  if char in freq:
    freq[char]+=1
  else:
    freq[char]=1
print(freq)
```

# **program to find the first non-repeating character in a string**

```python
s = "aabbcde"
freq = {}

# Count frequency
for char in s:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(freq)

# Find first non-repeating character
for ch in s:
    if freq[ch] == 1:
        print("First non-repeating character:", ch)
        break
else:
    print("No non-repeating character found")
```

# **program to find two strings are anagrams**

```python
s1 = "listen"
s2 = "silent"

if(sorted(s1) == sorted(s2)):
  print("yes it's an anagram")
else:
  print("no it's not an anagram")
```

# **program to remove duplicate characters from a string**

```python
str1 = "aacbdcdedadxdc"
print("original string : ",str1)
str2 = ""
for i in str1:
  if(i not in str2):
    str2+=i
print("string without duplicates : ",str2)
```
