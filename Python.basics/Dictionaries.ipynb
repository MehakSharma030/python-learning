# **Creating a Dictionary**

```python
# dictionary with a single record
student = {
    "Name " : "rohan",
    "Age " : 21,
    "Course ": "B.Tech",
    "Dept ":"CSE"
}
student
```

```python
# creating an empty dictionary
s = {}
print( type (s) )
```

```python
# dictionary with multiple records
student_record = {
    101 : {
        "Name " : "rohan",
        "Age " : 21,
        "Course ": "B.Tech",
        "Dept ":"CSE"
    },
    102 : {
        "Name " : "Arushi",
        "Age " : 20,
        "Course ": "B.Tech",
        "Dept ":"CSE"
    },
    103 : {
        "Name " : "Mehak",
        "Age " : 20,
        "Course ": "B.Tech",
        "Dept ":"CSE"
    }
}

print( student_record )
print( student_record[102]["Name "] )
print(student_record[103]["Dept "] )
```

```python
# creating a dictionary using a dictionary constructor
student = dict(name="Mehak", age=20, course="B.Tech")
print(student)
```

# Accessing Dictionary Elements

```python
student = {
    "name": "Arushi",
    "age": 20
}

print( student["name"] )
print( student.get("age") )
```

# **dictionary functions**

```python
student = {
    "name": "Arushi",
    "age": 20,
    "branch":"CSE"
}

print(student)

# updating the value of age from 20 to 21
student["age"] = 21
student.update( {"branch" : "AIML"} )
print(student)
```

```python
# keys() --> returns all keys of the dictionary
student = {
    "name": "Arushi",
    "age": 20,
    "course": "B.Tech",
    "dept ": "CSE"
}

print("keys() : ", student.keys() )
```

```python
# values() --> returns all values of the dictionary
student = {
    "name": "Arushi",
    "age": 20,
    "course": "B.Tech",
    "dept ": "CSE"
}

print("values() : ", student.values() )
```

```python
# items() --> returns all key value pairs of the dictionary in the form of a list of tuple
student = {
    "name": "Arushi",
    "age": 20,
    "course": "B.Tech",
    "dept ": "CSE"
}

print("items() : ", student.items() )
```

```python
# get() --> Returns the value for a key
student = {
    "name": "Arushi",
    "age": 20,
    "course": "B.Tech",
    "dept ": "CSE"
}

print("get() : ",student.get("name"))
```

```python
# pop() --> Removes and returns the value of a specified key
student = {
    "name": "Arushi",
    "age": 20,
    "course": "B.Tech",
    "dept ": "CSE"
}

print("pop() : ",student.pop("age"))
```

```python
# popitems() --> Removes and returns the last inserted item
student = {
    "name": "Arushi",
    "age": 20,
    "course": "B.Tech",
    "dept ": "CSE",
}

print("popitem() : ",student.popitem())
```

```python
# clear() --> Removes all items from the dictionary
student = {
    "name": "Arushi",
    "age": 20,
    "course": "B.Tech",
    "dept ": "CSE"
}

print(student)
student.clear()
print(student)
```

```python
# copy() --> Returns a shallow copy of the dictionary
student = {
    "name": "Arushi",
    "age": 20,
    "course": "B.Tech",
    "dept ": "CSE"
}

print("copy() : ",student.copy())
```

```python
# setdefault() --> Returns the value of a key if it exists ; inserts key if not present
student = {
   "name": "Arushi",
    "age": 20,
    "course": "B.Tech",
    "dept ": "CSE"
}

print("setdefault() : ",student.setdefault("name","arohi"))
```

```python
# fromkeys() --> Creates a new dictionary with specified keys and value
keys = ['a', 'b', 'c']

d = dict.fromkeys(keys,1)

print(d)
```
