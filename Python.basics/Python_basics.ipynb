# **DATA TYPES**

```python
# Different Data Types in Python

# Integer
a = 10

# Float
b = 10.5

# Complex
c = 2 + 3j

# String
d = "Hello"

# List
e = [1, 2, 3, 4]

# Tuple
f = (10, 20, 30)

# Set
g = {1, 2, 3, 4}

# Dictionary
h = {"name": "Mhk", "age": 20}

# Boolean
i = True

# None Type
j = None


# Printing values and their datatypes

print(a, ":", type(a))
print(b, ":", type(b))
print(c, ":", type(c))
print(d, ":", type(d))
print(e, ":", type(e))
print(f, ":", type(f))
print(g, ":", type(g))
print(h, ":", type(h))
print(i, ":", type(i))
print(j, ":", type(j))
```

# **OPERATORS**

```python
# Different Operators in Python

a = 10
b = 3


# Arithmetic Operators

print("Arithmetic Operators")
print("a + b =", a + b)
print("a - b =", a - b)
print("a * b =", a * b)
print("a / b =", a / b)
print("a // b =", a // b)
print("a % b =", a % b)
print("a ** b =", a ** b)

print()


#  Comparison Operators

print("Comparison Operators")
print("a == b :", a == b)
print("a != b :", a != b)
print("a > b :", a > b)
print("a < b :", a < b)
print("a >= b :", a >= b)
print("a <= b :", a <= b)

print()


# Assignment Operators

print("Assignment Operators")

x = 5

x += 2
print("x += 2 :", x)

x -= 1
print("x -= 1 :", x)

x *= 3
print("x *= 3 :", x)

x /= 2
print("x /= 2 :", x)

print()


#  Logical Operators

print("Logical Operators")

p = True
q = False

print("p and q :", p and q)
print("p or q :", p or q)
print("not p :", not p)

print()


#  Bitwise Operators

print("Bitwise Operators")

print("a & b =", a & b)
print("a | b =", a | b)
print("a ^ b =", a ^ b)
print("~a =", ~a)
print("a << 1 =", a << 1)
print("a >> 1 =", a >> 1)

print()


#  Membership Operators

print("Membership Operators")

lst = [1,2,3,4,5]

print("2 in lst :", 2 in lst)
print("10 not in lst :", 10 not in lst)

print()


#  Identity Operators

print("Identity Operators")

m = [1,2,3]
n = m
k = [1,2,3]

print("m is n :", m is n)
print("m is k :", m is k)
print("m is not k :", m is not k)
```

# **VARIABLES**

```python

# Integer Variable
a = 10

# Float Variable
b = 10.5

# String Variable
name = "Mhk"

# Boolean Variable
is_student = True


# Printing Variables

print("Integer value :", a)
print("Float value :", b)
print("Name :", name)
print("Student :", is_student)

print()

# Printing Datatypes

print("Datatype of a :", type(a))
print("Datatype of b :", type(b))
print("Datatype of name :", type(name))
print("Datatype of is_student :", type(is_student))
```

# **CONDITIONAL STATEMENTS**

```python
# Conditional Statements

age = 18

# if statement
if age > 18:
    print("You are an adult")

# if-else statement
if age >= 18:
    print("You can vote")
else:
    print("You cannot vote")

# if-elif-else statement
marks = 75

if marks >= 90:
    print("Grade A")

elif marks >= 70:
    print("Grade B")

elif marks >= 50:
    print("Grade C")

else:
    print("Fail")
```
