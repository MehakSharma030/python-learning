```python
# set_context() -> it is used to change the scaling/size of the entire plot including font size , title size , line width , marker size , axis label size

# | Context    | Use                            |
# | -----------| ------------------------------ |
# |  paper     | small plots for papers         |
# |  notebook  | default size                   |
# |  talk      | larger plots for presentations |
# |  poster    | very large plots               |
```

```python
sns.set_style("white")
sns.set_context("poster")
sns.scatterplot(data = penguins , x = "flipper_length_mm" , y = "body_mass_g" , hue = "species"	)
```

PAIR GRID

```python
# It's a subplot grid for plotting pairwise relationships in a dataset
graph = sns.PairGrid(data = penguins , hue = "sex" , palette = "Set2")
graph.map_upper(sns.scatterplot)  # upper triangle
graph.map_lower(sns.kdeplot)      # lower triangle
graph.map_diag(sns.histplot )     # diagonal elements

# to add the legend
graph.add_legend()
```
