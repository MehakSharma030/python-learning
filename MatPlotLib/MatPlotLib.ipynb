2. bivariate : numerical-categorial

BOX PLOT

```python
hr_sal = df [ df["Dept"] == "HR" ] ["Salary"]
it_sal = df [ df["Dept"] == "IT" ] ["Salary"]
fin_sal = df [ df["Dept"] == "Finance" ] ["Salary"]
```

```python
plt.boxplot( [hr_sal , it_sal , fin_sal] , labels = [ "HR" , "IT" , "Finance" ])
plt.show()
```

Pie Chart

```python
salary_by_dept = df.groupby("Dept")["Salary"].mean()
salary_by_dept
```

```python
plt.pie(salary_by_dept , labels = salary_by_dept.index , autopct = "%1.2f%%" , explode = [0.1,0.2,0.1] )
plt.show()
```

BAR CHART

```python
hr_mean = sum(hr_sal) / len(hr_sal)
it_mean = sum(it_sal) / len(it_sal)
fin_mean = sum(fin_sal) / len(fin_sal)
```

```python
plt.bar( ["HR" , "IT" , "Finance"] , [hr_mean , it_mean , fin_mean ] , color = [ "green" , "red" , "black" ] )
plt.show()
```

# **MULTIVARIATE ANALYSIS**

```python
df["Experience"] = [
    1, 2, 5, 3, 7,
    4, 6, 2.5, 8, 10,
    1, 6.5, 9, 3, 12.5,
    4, 9.7, 6, 2, 7.5
]
df
```

BUBBLE PLOT

```python
plt.scatter( df["Age"] , df["Salary"] , s = df["Experience"]*20 , color = "yellow" , edgecolor = "black" )
plt.show()
```

```python
plt.scatter( df["Age"] , df["Salary"] , s = df["Experience"]*20 , color = "yellow" , edgecolor = "black" )
plt.title("Age VS Salary Vs Experience")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid()
plt.show()
```

2. 2-numerical and 1-categorial data

```python
plt.scatter(
    df["Age"],
    df["Salary"],
    c = df["Dept"].map({
        'HR': "yellow",
        'IT': "black",
        'Finance': "red"
    })
)

plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age Vs Salary Vs Dept")
plt.show()
```

```python
color = {"HR" : "yellow" , "IT" : "black" , "Finance" : "red"}
for dept,color in color.items():
    dept_data = df[df["Dept"] == dept]
    plt.scatter(dept_data["Age"] , dept_data["Salary"] , label = dept , color = color)
    plt.legend()
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age Vs Salary Vs Dept")
plt.show()
```

# **`OBJECT ORIENTED API`**

```python
fig,axs = plt.subplots(1,3 , figsize = (15,5) )

# 1. LINE PLOT
axs[0].plot(sort_age["Age"] , df["Salary"] , color = "red" , marker = "*" , linewidth = '2')
axs[0].grid()
axs[0].set_title("line plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel("Salary")

# 2. HISTOGRAM
axs[1].hist(df["Salary"] , bins = 5 , color = "skyblue")
axs[1].grid()
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")

# 3. BOX PLOT
axs[2].boxplot( [hr_sal , it_sal , fin_sal] , labels = [ "HR" , "IT" , "Finance" ])
axs[2].grid()
axs[2].set_title("Box Plot")
axs[2].set_xlabel("Dept")

plt.savefig("multiplePlots.png")
plt.show()
```

# **3D PLOT**

```python
data3 = {
    "Name": [
        "Alice", "Bob", "Charlie", "David", "Eva",
        "Frank", "Grace", "Henry", "Ivy", "Jack"
    ],

    "Age": [
        22, 28, 25, 30, 27,
        35, 24, 32, 26, 29
    ],

    "Experience": [
        1, 5, 3, 7, 4,
        10, 2, 8, 3, 6
    ],

    "Salary": [
        25000, 50000, 35000, 70000, 45000,
        90000, 30000, 80000, 40000, 60000
    ]
}

import pandas as pd

df2 = pd.DataFrame(data3)

print(df2)

fig = plt.figure()

ax = plt.axes(projection="3d")

ax.scatter(df2["Age"],
           df2["Salary"],
           df2["Experience"])

ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_zlabel("Experience")

plt.show()
```
